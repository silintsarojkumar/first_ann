## Import Library

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
```

## Load the Dataset

```python
df = pd.read_csv("Churn_Modelling.csv")
```

```python
df.head()
```

## Import Some More Library

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler,LabelEncoder
import pickle
```

```python
## Preprocesss the data

df = df.drop(['RowNumber','CustomerId','Surname'],axis=1)
df.head()
```

```python
## Encode Categrical variable
label_encoder_gender = LabelEncoder()
df['Gender'] = label_encoder_gender.fit_transform(df['Gender'])
```

```python
df
```

```python
df['Geography'].unique()
```

```python
from sklearn.preprocessing import OneHotEncoder
onehot_geo = OneHotEncoder()
geo_onehot = onehot_geo.fit_transform(df[['Geography']])
```

```python
onehot_geo.get_feature_names_out(['Geography'])
```

```python
geo_onehot = geo_onehot.toarray()
```

```python
geo_onehot_df = pd.DataFrame(geo_onehot,columns=onehot_geo.get_feature_names_out(['Geography']))
```

```python
geo_onehot_df
```

```python
df = pd.concat([df.drop('Geography',axis=1),geo_onehot_df],axis=1)
```

```python
df.head()
```
